```python
### Answer 1
def highest_frequency_word_length(string):
    words = string.split()
    frequency = {}

    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    max_frequency = max(frequency.values())
    max_frequency_words = [word for word, freq in frequency.items() if freq == max_frequency]
    longest_word_length = max(len(word) for word in max_frequency_words)

    return longest_word_length

# Test case 1
string1 = "write write write all the number from from from 1 to 100"
print(highest_frequency_word_length(string1))

# Test case 2
string2 = "hello world hello world hello world"
print(highest_frequency_word_length(string2))

# Test case 3
string3 = "apple apple banana banana cherry cherry cherry"
print(highest_frequency_word_length(string3))
```

The maximum frequency is determined using the max() function on the dictionary values.

```python
### Answer 2
def is_valid_string(s):
    frequency = {}
    
    for char in s:
        frequency[char] = frequency.get(char, 0) + 1
    
    most_common_freq = max(frequency.values())
    
    diff_freq_count = sum(freq != most_common_freq for freq in frequency.values())

    if diff_freq_count <= 1:
        return "YES"
    else:
        return "NO"

# Test case 1
s1 = "abc"
print(is_valid_string(s1))  # Output: YES

# Test case 2
s2 = "abcc"
print(is_valid_string(s2))  # Output: NO

# Test case 3
s3 = "aabbcc"
print(is_valid_string(s3))  # Output: YES

# Test case 4
s4 = "aabbccc"
print(is_valid_string(s4))  # Output: YES
```
